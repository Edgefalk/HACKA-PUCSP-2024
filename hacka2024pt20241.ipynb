```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("train_df.tsv", sep='\t')


palavras_chave = ['Pobreza', 'Fome', 'Saude', 'Educacao', 'saneamento', 'energia', 'clima',]


contagem_palavras = {palavra: 0 for palavra in palavras_chave}


for titulo in df['titulo']:
    for palavra in palavras_chave:
        if palavra.lower() in titulo.lower():
            contagem_palavras[palavra] += 1


df_contagem = pd.DataFrame.from_dict(contagem_palavras, orient='index', columns=['Frequência'])


df_contagem = df_contagem.sort_values(by='Frequência', ascending=False)


plt.figure(figsize=(10, 6))
df_contagem.plot(kind='bar', color='skyblue')
plt.title('Frequência das Palavras-Chave nos Títulos')
plt.xlabel('Palavra-Chave')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
